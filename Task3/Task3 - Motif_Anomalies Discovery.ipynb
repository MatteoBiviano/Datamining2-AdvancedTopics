```python
#from matrixprofile import compute
# matrixprofile library as matrixprofile_ts
#from matrixprofile_ts import *
#import matrixprofile as matp
from matrixprofile import *
```

```python
def matrixProfile_stomp(ts, w, name):
    # ts = Temperature_2["Temperature"].values
    # Specifichiamo una finestra
    #w = 120
    mp, mpi = matrixProfile.stomp(ts, w)
    # Stomp  uno dei modi possibili con cui si pu estrarre la matrix profile
    plt.title("Matrix Profile" + str(name) + " - w = " + str(w))
    plt.plot(mp)
    plt.show()
    return mp, mpi
```

# Motifs Discovery

Proviamo un giorno feriale (Giovedi) e Temperature con finestra w =45

```python
# Ora troviamo i migliori motifs. Passiamo la time series, la matrix profile e specifichiamo 
# quanti motifs vogliamo

w=25
ts = Temperature_Venerdi["Temperature"].values
mp, mpi = matrixProfile_stomp(ts, w, "Temperature")
mo, mod  = motifs.motifs(ts, (mp, mpi), max_motifs=5)
```

```python
mo
```

```python
mod
```

```python
plt.plot(ts)
colors = ['k', 'g', 'r', 'y', 'b'][:len(mo)]
for m, d, c in zip(mo, mod, colors):
    for i in m:
        m_shape = ts[i:i+w]
        plt.plot(range(i,i+w), m_shape, color=c, lw=3)

plt.grid()
plt.show()

tserie = Temperature_Venerdi["Temperature"]
index = pd.date_range(start=tserie.index[0], end=tserie.index[len(ts)-1], freq='1Min')
tserie = pd.Series(tserie.values[:-1], index)
plt.plot(tserie)
plt.xticks(rotation=35)
plt.show()
```

Parameters  
- mp: matrix profile numpy array
- k: the number of discords to discover
- ex_zone: the number of samples to exclude and set to Inf on either side of a found discord    

Returns 
 - a list of indexes represent the discord starting locations. MaxInt indicates there were no more discords that could be found due to too many exclusions or profile being too small. Discord start indices are sorted by highest matrix profile value.
