Acquisisco la data dai dataset iniziali in tre dataframe separati

```python
# Dataset con attributi originali
# Dataset from 2015-02-04 17:51:00 to 2015-02-10 09:33:00
df3 = pd.read_csv("datatraining.txt")
# Dataset from 2015-02-02 14:19:00 to 2015-02-04 10:43:00
df1 = pd.read_csv("datatest.txt")
# Dataset from 2015-02-11 14:48:00 to 2015-02-18 09:19:00
df2 = pd.read_csv("datatest2.txt")
```

```python
Time_2 = df3["date"]
Time_3 = df2["date"]
Time_1 = df1["date"]
# Dimensioni dei dataframe
dim1 = len(Time_1)
dim2 = len(Time_2)
dim3 = len(Time_3)
time_columns = pd.concat([Time_1, Time_2, Time_3], ignore_index=True)
```

```python
Time_1 = time_columns[0:dim1]
Time_2 = time_columns[dim1:(dim1+dim2)]
Time_3 = time_columns[(dim1+dim2):(dim1+dim2+dim3)]
```

```python
dat_primo_giorno_giovedi = []
for i in Time_2:
    if i >= "2015-02-05 00:00:00" and i < "2015-02-06 00:00:00":
        dat_primo_giorno_giovedi.append(i)
        
dat_primo_giorno_venerdi = []
for i in Time_2:
    if i >= "2015-02-06 00:00:00" and i < "2015-02-07 00:00:00":
        dat_primo_giorno_venerdi.append(i)
        
dat_primo_giorno_sabato = []
for i in Time_2:
    if i >= "2015-02-07 00:00:00" and i < "2015-02-08 00:00:00":
        dat_primo_giorno_sabato.append(i)
        
dat_primo_giorno_domenica = []
for i in Time_2:
    if i >= "2015-02-08 00:00:00" and i < "2015-02-09 00:00:00":
        dat_primo_giorno_domenica.append(i)
        
dat_primo_giorno_lunedi = []
for i in Time_2:
    if i >= "2015-02-09 00:00:00" and i < "2015-02-10 00:00:00":
        dat_primo_giorno_lunedi.append(i)
```

Ci salviamo 5 giorni della seconda settimana da giovedì a lunedì

```python
dat_secondo_giorno_giovedi = []
for i in Time_3:
    if i >= "2015-02-12 00:00:00" and i < "2015-02-13 00:00:00":
        dat_secondo_giorno_giovedi.append(i)
        
dat_secondo_giorno_venerdi = []
for i in Time_3:
    if i >= "2015-02-13 00:00:00" and i < "2015-02-14 00:00:00":
        dat_secondo_giorno_venerdi.append(i)
        
dat_secondo_giorno_sabato = []
for i in Time_3:
    if i >= "2015-02-14 00:00:00" and i < "2015-02-15 00:00:00":
        dat_secondo_giorno_sabato.append(i)
        
dat_secondo_giorno_domenica = []
for i in Time_3:
    if i >= "2015-02-15 00:00:00" and i < "2015-02-16 00:00:00":
        dat_secondo_giorno_domenica.append(i)
        
dat_secondo_giorno_lunedi = []
for i in Time_3:
    if i >= "2015-02-16 00:00:00" and i < "2015-02-17 00:00:00":
        dat_secondo_giorno_lunedi.append(i)
```
