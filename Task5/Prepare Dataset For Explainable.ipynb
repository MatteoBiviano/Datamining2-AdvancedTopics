```python
# Dataset con attributi originali
df = pd.read_csv("../dataset_All_normalized.csv")
class_name = 'Occupancy'
```

```python
Temperature = df["Temperature"]
Light = df["Light"]
CO2 = df["CO2"]
HumidityRatio = df["HumidityRatio"]
Day = df["Day"]
Afternoon = df["Afternoon"]
Evening = df["Evening"]
Night = df["Night"]
Weekday = df["Weekday"]
Weekend = df["Weekend"]
Occupancy = df["Occupancy"]
```

```python
Time_Span = ["OO" for i in range(0,len(Day))]
for i in range(0, len(Day)):
    if Day[i] == 1:
        Time_Span[i] = "DA"
    if Afternoon[i] == 1:
        Time_Span[i] = "AF"
    if Evening[i] == 1:
        Time_Span[i] = "EV"
    if Night[i] == 1:
        Time_Span[i] = "NI"
```

```python
Week_Span = ["OO" for i in range(0,len(Weekday))]
for i in range(0, len(Weekday)):
    if Weekday[i] == 1:
        Week_Span[i] = "WD"
    if Weekend[i] == 1:
        Week_Span[i] = "WE"
```

```python
d = {'Temperature': Temperature, 'Light': Light, 'CO2': CO2,
        'HumidityRatio': HumidityRatio, 'TimeSpan': Time_Span, 'WeekSpan': Week_Span, 
            'Occupancy': Occupancy}
dataframe = pd.DataFrame(data=d)
```
